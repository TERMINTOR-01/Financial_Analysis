# Data Cleanning

## Drop Un-Wanted Columns

```python
df.drop(
    columns=[
        "img_link",
        "product_link",
        "review_content",
        "review_title",
        "review_id",
        "user_name",
        "user_id",
        "about_product"
    ],inplace=True
)
```

```python
df[df["rating_count"].isnull()]
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.isnull().sum()
```

## Fix data types

##### Datatype change Required

1.discounted_price->float

2.actual_price->float

3.discount_percentage->int

4.rating->float
5.rating_count->int

```python
# Price
price_columns = ["discounted_price", "actual_price"]

df[price_columns] = (
    df[price_columns]
    .replace({"₹": "", ",": ""}, regex=True)
    .astype(float)
)

print(df)
```

```python
df
```

```python
df.discount_percentage = (
    df.discount_percentage
    .replace({"%": "", ",": ""}, regex=True)
    .astype(float)
)
```

```python
df
```

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
```

```python
df.rating_count = (
    df.rating_count
    .replace({"%": "", ",": ""}, regex=True)
    .astype(float)
)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

## string parsing/splitting

```python
## Category fix
df["category"] = df["category"].str.split("|").str[0]
```

```python
df
```

# Data Transformation

## Feature engineering

```python
df["Revenue_Potential"]=df["discounted_price"]*df["rating_count"]
```

```python
df
```

```python
df['absolute_discounted_amount']=df['actual_price']-df['discounted_price']
```

```python
df
```
